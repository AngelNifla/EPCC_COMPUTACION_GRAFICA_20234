#*A. FUNCION*


```python
def rotar_imagen(Imagen_nombre):
  # Cargar la imagen
  image = cv2.imread(Imagen_nombre, cv2.IMREAD_COLOR)
  if image is None:
      raise ValueError("Error\n")

  gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)  # Convertir la imagen a escala de grises
  _, binary = cv2.threshold(gray, 150, 255, cv2.THRESH_BINARY_INV)  # Uumbral binario
  contours, _ = cv2.findContours(binary, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE) # Encontrar los contornos


  largest_contour = max(contours, key=cv2.contourArea)  # Halla el rectángulo donde se encuentra el texto
  rect = cv2.minAreaRect(largest_contour)

  # Hallando el angulo de inclinación
  angulo = rect[2]

  print(angulo)
  if angulo < -45:
      angulo = 90 + angulo
  elif angulo == 90:
      angulo = angulo - 180
  elif angulo > 45:
      angulo = angulo - 90
  else:
      angulo = angulo - 90

  #angulo = (angulo - 180)
  print("->")
  print(angulo)

  # Para las dimenciones de la imagen
  (H, W) = image.shape[:2]
  centro = (W // 2, H // 2)


  M = cv2.getRotationMatrix2D(centro, angulo, 1.0) # Calculando la matriz de rotación
  imagen_rotada = cv2.warpAffine(image, M, (W, H), flags=cv2.INTER_LINEAR, borderMode=cv2.BORDER_CONSTANT, borderValue=(0, 0, 0)) # Rotando la imagen, sin cambiar los fondos.

  return imagen_rotada
```

##*Imagen de prueba*

```python
Imagen_nombre = 'image1.png'

Nueva_Imagen = rotar_imagen(Imagen_nombre)

# Se muestra la imagen original y la imagen rotada
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.title('Original')
plt.imshow(cv2.cvtColor(cv2.imread(Imagen_nombre), cv2.COLOR_BGR2RGB))
plt.axis('off')

plt.subplot(1, 2, 2)
plt.title('Rotada')
plt.imshow(cv2.cvtColor(Nueva_Imagen, cv2.COLOR_BGR2RGB))
plt.axis('off')

plt.show()
```
